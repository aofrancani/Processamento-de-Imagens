# Leitura de Imagens 

Neste notebook eu vou explorar conceitos básicos de leitura de imagens no Python 3 usando algumas bibliotecas comuns, sendo elas OpenCV, PIL e matplotlib.

##  Usando OpenCV

A primeira biblioteca que vou utilizar é a OpenCV, muito utilizada em problemas de visão computacional. A imagem será lida com a função ``cv2.imread()``.

````python
import cv2
import os

# path to the desired image
image_name = 'doguinho.jpg'
image_path = os.path.join(os.getcwd(), 'images', image_name)

# read image
img = cv2.imread(image_path)
````

````python
# print image info
print('Image shape: ', img.shape)
print('Image size:', img.size)
print('Image type: ', img.dtype)
print('Max value = %d   Min value = %d ' %(img.max(), img.min()))
````

Podemos observar que a imagem é um array de dimensões $1280 \times 960$ e $3$ canais (RGB). Os valores dos pixels são números inteiros entre $0$ e $255$, ou seja, são codificados com $8$ bits (uint8), podendo assumir $2^8 = 256$ valores inteiros.

### Visualizando a imagem

Agora, vamos visualizar a imagem. Você pode utilizar a função ``cv2.imshow()`` da seguinte forma:

```python
# display image
cv2.imshow('Nome da imagem', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Isso fará com que abra uma janela no seu computador com a respectiva imagem em na sua variável *_img_*. Entretanto, eu não costumo muito usar essa função, então vou mostrar a imagem usando a biblioteca *_matplotlib_*:

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.imshow(img);
plt.axis('off');
````

Interessante... a imagem não foi mostrada como esperado. Isso ocorre porque o OpenCV lê os canais na ordem BGR (Blue, Green, Red) e não RGB (Red, Green, Blue). Logo, para usar a biblioteca do *_matplotlib_* para mostrar a imagem corretamente, devemos primeiro converter a representação de BGR para RGB.

````python
# reads image and converts from BGR to RGB
img_rgb = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12,8))
plt.imshow(img_rgb);
plt.axis('off');
````

Agora sim dá para ver como ele é fofinho!

### Podemos visualizar os canais separadamente?

Sim! Uma das maneiras de fazer isso é colocar zeros nos demais canais e continuar com os valores do canal de cor desejado. Para isso, vou criar um numpy array com zeros e copiar os valores dos pixels no canal desejado.

````python
import numpy as np

# red channel
img_red = np.zeros(img_rgb.shape, 'uint8') # create array with zeros
img_red[:,:,0] = img_rgb[:,:,0]            # get pixel values of red channel

# green channel
img_green = np.zeros(img_rgb.shape, 'uint8') # create array with zeros
img_green[:,:,1] = img_rgb[:,:,1]            # get pixel values of green channel

# blue channel
img_blue = np.zeros(img_rgb.shape, 'uint8') # create array with zeros
img_blue[:,:,2] = img_rgb[:,:,2]            # get pixel values of blue channel
````

````python
# show RGB images 
plt.figure(figsize=(12,8))
plt.subplot(1,3,1)
plt.imshow(img_red)
plt.title('Red')
plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(img_green)
plt.title('Green')
plt.axis('off')
plt.subplot(1,3,3)
plt.imshow(img_blue)
plt.title('Blue')
plt.axis('off');
````

### Como ver a imagem em tons de cinza (grayscale)?

Agora que vimos a imagem em seus canais RGB, há uma maneira muito rápida e simples de ler a imagem diretamente em níveis de cinza. Pretendo fazer um notebook mais para frente sobre as escalas de cores e a conversão entre elas, mas no momento vamos fazer de maneira mais simples. Para isso, basta colocar um segundo parâmetro '0' na função ``cv2.imread()`` como visto [aqui neste link](https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_gui/py_image_display/py_image_display.html):

```python
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
                OU
img = cv2.imread(image_path, 0)
```

Feito isso, devemos no *_matplotlib_* também especificar o *_colormap_* como 'gray'.

````python
# loads image as grayscale
img_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(12,8))
plt.imshow(img_gray, cmap='gray'); # colormap as grayscale
plt.axis('off');
````

Observe que agora a imagem não possui 3 canais, mas apenas 1 canal:

````python
print('Image shape: ', img_gray.shape)
````

### Salvando a imagem

Por fim, podemos salvar as imagens usando a função ``cv2.imwrite()`` do próprio OpenCV da seguinte forma:

```python
cv2.imwrite('nome_da_imagem.jpg', img)
```

O *_matplotlib_* também pode ser utilizado para salvar as figuras com [``plt.imsave()``](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imsave.html) conforme a seguir:

```python
plt.imsave('nome_da_imagem.jpg', img)
```

## Usando PIL

A biblioteca que será utilizada nesta seção é a [PIL](https://pillow.readthedocs.io/en/stable/reference/Image.html) (Python Imaging Library). Assim, vamos importar a classe ``Image`` de PIL e carregar a imagem com o método ``open()``:

````python
from PIL import Image

# read image
img = Image.open(image_path)
````

````python
display(img.size)
img.resize((240, 320)) # the normal size was too big to display
````

Se você quiser abrir a imagem em uma outra janela, você pode usar o método

```python
img.show()
```

Note que agora a variável *_img_* é um objeto do tipo ``Image``. Portanto, há diversos métodos desse objeto que podem ser utilizados, podendo serem vistos [aqui](https://pillow.readthedocs.io/en/3.0.x/reference/Image.html#functions).

Supondo que você tenha uma imagem do tipo array e queira convertê-la para o objeto ``Image``, você pode utilizar a função ``Image.fromarray()``:

```python
img_array = np.asarray(img)       # convert Image object to numpy array
img2 = Image.fromarray(img_array) # convert numpy array to Image object
```

### Lendo imagem em níveis de cinza

Há uma função ``Image.convert()`` com diversos [modos](https://pillow.readthedocs.io/en/stable/handbook/concepts.html#concept-modes). A conversão para o formato *_grayscale_* é dada por 'L': 

````python
img_gray = img.convert('L')
img_gray.resize((240, 320)) # the normal size was too big to display
````

### Salvando a imagem

Caso você queira salvar a sua imagem modificada, como por exemplo a imagem em níveis de cinza, basta utilizar a função

```python
img_gray.save('nome_da_imagem.jpg')
```
