# Sistemas de Cores

Neste notebook vamos dar uma olhada bem rápida nos sistemas de cores para representar uma imagem, sendo eles o RGB, CMY, níveis de cinza (_grayscale_), HSV e YCbCr.

## 1. RGB

O sistema RGB possui esse nome graças a suas cores _red_ (vermelho), _green_ (verde) e _blue_ (azul). Esse modelo de cores é muito utilizado em monitores de televisão e telas de celulares, tanto que você pode testar os LEDs de cada uma dessas cores individualmente em seu celular. 

Como vimos no notebook de [leitura de imagens](https://github.com/aofrancani/Processamento-de-Imagens/blob/main/1%20-%20Leitura%20de%20Imagens.ipynb), a imagem lida possui 3 canais (RGB ou BGR no OpenCV), ou seja, cada cor de píxel é definida por 3 valores inteiros. Por exemplo, se o sistema for codificado com 8 bits, o que é usual, os píxels são valores inteiros entre 0 e 255 ($2^8 = 256$). Assim, pensando no sistema de coordenadas tri-dimensional, temos $256 \times 256 \times 256 = 16.777.216$ cores possíveis. As cores mais simples de saber os valores nesse sistema são:

     - vermelha: (R, G, B) = (255, 0, 0);
     - verde: (R, G, B) = (0, 255, 0);
     - azul: (R, G, B) = (0, 0, 255);
     - branca: (R, G, B) = (255, 255, 255);
     - preta: (R, G, B) = (0, 0, 0).

Uma maneira simples de pensar é que na cor branca, as 3 cores RGB estão "ativadas", misturando as cores como no disco de Newton que a gente já estudou na escola. Na cor preta, nenhum canal está ativado. Na vermelha, apenas o canal vermelho está "ativado", e assim por diante. O valor 255 indica o valor máximo de "ativação", e outros valores darão outras intensidades de cores.


O sistema de cores RGB é aditivo, ou seja, as demais cores são representadas pela soma de cores, como no GIF abaixo:  

<img src="https://upload.wikimedia.org/wikipedia/commons/8/89/RGB_colors.gif" width="200" height="100" alt="Sistema de cores RGB">

As demais cores podem ser vistas nos pontos do sistema de coordenadas 3D abaixo:

<img src="https://upload.wikimedia.org/wikipedia/commons/a/af/RGB_color_solid_cube.png" width="300" height="200"  alt="RGB cube">


Vamos ver agora um exemplo bem rápido em Python variando apenas o valor do pixel do canal R (vermelho).

```python
import numpy as np
import matplotlib.pyplot as plt

# create red image array with dimension 30x30x3
img_red = np.zeros((30,30,3), 'uint8')
img_red[:,:,0] = 1;

# plot red images with different shades
plt.figure(figsize=(12,8))
plt.subplot(1,3,1)
plt.imshow(255*img_red)  # red channel with 255
plt.title('(R,G,B) = (255, 0, 0)')
plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(180*img_red)  # red channel with 180
plt.title('(R,G,B) = (180, 0, 0)')
plt.axis('off')
plt.subplot(1,3,3)
plt.imshow(100*img_red)  # red channel with 100
plt.title('(R,G,B) = (100, 0, 0)')
plt.axis('off');
```

## 2. CMY ou CMYK

Observando o GIF do RGB, pode-se ver que a mistura das cores básicas do RGB formam as cores ciano, magenta e amarelo. Assim, as cores também podem ser representadas por esse sistema de cores CMY (Cyan, Magenta, Yellow) ou CMYK (Cyan, Magenta, Yellow, Black). Esse sistema de cores é muito utilizado em tintas de impressoras. O CMYK contém a cor preta pois normalmente a mistura CMY não resulta perfeitamente na cor preta. Além disso, pensando em impressoras, a tinta preta ajudaria também na economia das tintas coloridas. O sistema CMY é do tipo subtrativo.

## 3. Níveis de Cinza

Agora, o sistema em níveis de cinza (ou tons de cinza) possui apenas um canal. Muitas vezes esse sistema é utilizado para visualizar a imagem em preto e branco. Como converter uma imagem com 3 canais em apenas 1 canal? Vamos dar uma estudada rápida.

### 3.1 Média dos Pixels

Será que se eu apenas tirar a média dos 3 canais para todos os pixels eu consigo ver a imagem em níveis de cinza? Vamos ver:

```python
import cv2
import os

image_path = os.path.join(os.getcwd(), 'images', 'doguinho.jpg') # path to the desired image
img = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB) # read image in RGB

# convert RGB to Grayscale using mean value
img_gray_average = np.mean(img, axis=2) # mean over the last dimension (channels)
img_gray_average = np.round(img_gray_average)

# show images
plt.figure(figsize=(8,6))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(img_gray_average, cmap='gray')
plt.title('Grayscale - average value')
plt.axis('off');
```

Parece que funcionou bem, não? Mas será que é feito assim mesmo no [OpenCV](https://docs.opencv.org/3.4/de/d25/imgproc_color_conversions.html)?

### 3.2 Conversão ajustada à visão humana <a id='grayscale'></a>

Ao longo dos estudos e pesquisas, foi identificado que o olho humano é mais sensível à banda da luz visível verde e menos sensível à banda azul. Assim, é conveniente aplicar uma média ponderada ajustando os pesos de acordo com a percepção humana. A equação que converte de RGB para níveis de cinza é

$$ Y_{gray} = 0,299 R + 0,587G + 0,114B $$.

Vamos ver isso em código e comparar visualmente com as imagens em nível de cinza da média simples e do OpenCV.


```python
img_gray_opencv = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE) #read image in grayscale

# RGB to Grayscale
Y = 0.299*img[:,:,0] + 0.587*img[:,:,1] + 0.114*img[:,:,2]
Y = np.round(Y) 

# show images
plt.figure(figsize=(16,12))
plt.subplot(1,3,1)
plt.imshow(img_gray_average, cmap='gray')
plt.title('Average')
plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(Y, cmap='gray')
plt.title('Weighted average')
plt.axis('off');
plt.subplot(1,3,3)
plt.imshow(img_gray_opencv, cmap='gray')
plt.title('Grayscale - OpenCV')
plt.axis('off');
```

As 3 imagens parecem muito parecidas. Entretanto, a primeira dada pela média simples aparenta um pouco mais escura que as demais. Como comparar então as imagens sem ser por inspeção visual? Há alguma métrica? Esse tema é algo que pretendo mostrar em outro notebook, mas vamos calcular um simples MAE (_mean absolute error_).

### MAE

Analisando a natureza das imagens, com valores entre 0 e 255, a maior média da diferença absoluta possível entre duas imagens é 255 (uma branca e outra preta). Assim, para expressar essa métrica em porcentagem, vou dividir essa diferença por 255.

```python
#MAE (mean absolute error)
MAE_average = np.mean(np.abs(img_gray_average - img_gray_opencv))/255
MAE_weighted = np.mean(np.abs(Y - img_gray_opencv))/255
print('MAE média: %.2f%%' %(100*MAE_average))
print('MAE ponderada: %.2f%%' %(100*MAE_weighted))
```

Realmente utilizando a média ponderada conseguimos a conversão de RGB para níveis de cinza idêntica ao do OpenCV, uma vez que houve 0% de erro nos valores dos pixels das imagens. 

## 4. HSI, HSL ou HSV

Um outro sistema de cores é o HSI, HSL ou HSV, que vem de matiz/coloração (_Hue_), saturação (_Saturation_), luminosidade (_Lightness_), intensidade (_Intensity_) e valor (_Value_). 

<a title="SharkD, CC BY-SA 3.0 &lt;https://creativecommons.org/licenses/by-sa/3.0&gt;, via Wikimedia Commons" href="https://commons.wikimedia.org/wiki/File:HSV_color_solid_cylinder.png"><img width="300" height="200" alt="HSV color solid cylinder" src="https://upload.wikimedia.org/wikipedia/commons/thumb/4/4e/HSV_color_solid_cylinder.png/512px-HSV_color_solid_cylinder.png"></a>

Vamos ver esse sistema de cores na nossa imagem do cachorrinho fofo em HSV. Você vai perceber que a intensidade ou valor é o mais próximo do nível de cinza. 

```python
img_hsv = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2HSV) # read image in HSV

# show images
plt.figure(figsize=(16,12))
plt.subplot(1,4,1)
plt.imshow(img)
plt.title('Original')
plt.axis('off')
plt.subplot(1,4,2)
plt.imshow(img_hsv[:,:,0], cmap='gray')
plt.title('Hue (H)')
plt.axis('off');
plt.subplot(1,4,3)
plt.imshow(img_hsv[:,:,1], cmap='gray')
plt.title('Saturation (S)')
plt.axis('off');
plt.subplot(1,4,4)
plt.imshow(img_hsv[:,:,2], cmap='gray')
plt.title('Value (V)')
plt.axis('off');
```

Maiores detalhes dessa conversão podem ser vistas na [conversão de cores do OpenCV](https://docs.opencv.org/3.1.0/de/d25/imgproc_color_conversions.html#color_convert_rgb_hsv). Cabe lembrar que esse sistema, assim como os outros, são alternativos ao sistema RGB. Dependendo da sua aplicação, pode ser que você queira implementar um sistema robusto a níveis de iluminação, então esse sistema de cores pode ser interessante de ser utilizado.


## 5. YUV ou YCbCr

Por fim, o último sistema de cores que vou mostrar aqui é o YUV ou YCbCr. Neste sistema, a imagem é representada por luminância e crominâncias. Esse sistema é utilizado, por exemplo, na codificação de imagens segundo o padrão JPEG. Nesse padrão, a informação de crominância é armazenada com menor resolução, uma vez que o olho humano é mais sensível à luminosidade do que crominância (há mais receptores para luminosidade do que para cores no olho humano).

A luminância (Y) é a imagem em níveis de cinza como vimos [aqui](#grayscale) , e os termos Cb e Cr são as crominâncias. A conversão de [RGB para YCbCr](https://docs.opencv.org/3.4/de/d25/imgproc_color_conversions.html) é dada por:

$$ Y = \alpha_R R + \alpha_GG + \alpha_B B, \\
C_b = \frac{1}{2(1-\alpha_B)} (B - Y) + \Delta, \\
C_r =  \frac{1}{2(1-\alpha_R)} (R - Y) + \Delta,$$

em que $\alpha_R = 0,299$, $\alpha_G = 0,587$, $\alpha_B = 0,114$ e $\Delta = 128$ para uma imagem de 8 bits.

Essa conversão irei implementar em Python apenas como exercício:

```python
# parameters
alpha_R = 0.299
alpha_G = 0.587
alpha_B = 0.114
delta = 128
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]

# RGB to YCbCr
Y = np.round(alpha_R*R + alpha_G*G + alpha_B*B)
Cb = (B-Y)/(2*(1-alpha_B)) + delta
Cr = (R-Y)/(2*(1-alpha_R)) + delta

# show images
plt.figure(figsize=(16,12))
plt.subplot(1,4,1)
plt.imshow(img)
plt.title('Original')
plt.axis('off')
plt.subplot(1,4,2)
plt.imshow(Y, cmap='gray')
plt.title('Y')
plt.axis('off');
plt.subplot(1,4,3)
plt.imshow(Cb, cmap='gray')
plt.title('Cb')
plt.axis('off');
plt.subplot(1,4,4)
plt.imshow(Cr, cmap='gray')
plt.title('Cr')
plt.axis('off');
```
